```python
def extract_data(sheet, countries, column_name, header_row=0):
    # Load correctly using detected header row
    df = pd.read_excel("/content/energy.xlsx", sheet_name=sheet, header=header_row)

    # Clean header names
    df.columns = df.columns.str.strip()

    # Identify the first column name again (it may not be the first visually)
    first_col = df.columns[0]

    # Rename the first column to 'countries'
    df.rename(columns={first_col: "countries"}, inplace=True)

    # Ensure index is set correctly
    df.set_index("countries", inplace=True)

    # Debug
    print("Columns:", df.columns)
    print("Index sample:", df.index[:5])

    # Extract specific countries
    df = df.loc[countries]

    # Rename the selected data column
    df.rename(columns={df.columns[0]: column_name}, inplace=True)

    return df

```

```python
extract_data('energy','Japan','Access to electricity (% of population)')
```
