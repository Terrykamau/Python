```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
x=np.linspace(0,10,100)
y=np.sin(x)
z=np.cos(x)
```

```python
print(x)
```

```python
print(y)
```

```python
print(z)
```

```python
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,z)
plt.show()
```

```python
plt.plot(x,y)
plt.xlabel('angle')
plt.ylabel('sine wave')
plt.title('sine wave')
plt.show()
```

```python
x=np.linspace(-10,10,20)
y=x**2
plt.plot(x,y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('quadratic')
plt.show()
```

```python
plt.plot(x,y,'r+')
plt.show()
```

```python
plt.plot(x,y,'g.')
plt.show()
```

```python
plt.plot(x,y,'bx')
plt.show()
```

```python
plt.plot(x,y,'ro')
plt.show()
```

```python
x=np.linspace(-5,5,50)
plt.plot(x,np.sin(x),'go')
plt.plot(x,np.cos(x),'b+')
plt.show()
```

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
languages= ['english','spanish','German','french']
people = [100,30,50,80]
ax.bar(languages,people)
plt.xlabel('languages')
plt.ylabel('people')
plt.show()
```

```python
w=plt.figure()
ax=w.add_axes([0,0,1,1])
languages= ['english','spanish','German','french']
people = [100,30,50,80]
ax.pie(people,labels=languages,autopct='%1.1f%%')
plt.show()
```

```python
x=np.linspace(0,10,30)
y=np.sin(x)
z=np.cos(x)
dia=plt.figure()
ax=dia.add_axes([0,0,1,1])
ax.scatter(x,y,color='red')
ax.scatter(x,z,color='blue')
plt.show()
```

```python
d=plt.figure()
ax=plt.axes(projection='3d')
z=20*np.random.random(100)
x=np.sin(z)
y=np.cos(z)
ax.scatter3D(x,y,z,c=z,cmap='Greens')
plt.show()
```
