```python
import pandas as pd
import numpy as np
```

```python
df=pd.DataFrame({
    'Date':pd.date_range(start='2025-01-10', periods=60),
    'Value':np.random.randint(0,100,size=60)
})
```

```python
print(df)
```

```python
df['year/month']=df['Date'].dt.strftime('%m/%y')
```

```python
print(df)
```

```python
df['week/year']=df['Date'].dt.strftime('%U/%y')
```

```python
print(df)
```

```python
df.groupby('week/year')['Value'].sum()
```

```python
df.groupby('week/year')['Value'].mean()
```

```python
df.groupby('week/year')['Value'].agg(['sum','mean'])
```
