```python
import pandas as pd
import numpy as np
```

```python
data={'name': ['teresiah','Njoki','kamau','wambui','Mary'],
      'country':['kenya','uganda',1,'djibouti',np.nan],
      'age':[20,30,40,np.nan,60]}
df=pd.DataFrame(data)
print(df)
```

```python
import statsmodels.api as sm
```

```python
df['country_encode']=df['country'].astype('category').cat.codes
print(df)
```

```python
x=df['country_encode']
y=df['age']
```

```python
x=sm.add_constant(x)
```

```python
model=sm.OLS(y, x).fit()
```

```python
model.summary
```

```python
df['name_encode']=df['name'].astype('category').cat.codes
print(df)
```

```python
x=df['name_encode']
y=df['age']
```

```python
x=sm.add_constant(x)
```

```python
model=sm.OLS(y, x).fit()
```

```python
model.summary()
```

```python
df1=pd.DataFrame({'ID':[1,2,3,4],
                  'Names':['james', 'john','Andrew','Matthew']})
```

```python
df2=pd.DataFrame({'ID':[2,3,4],
                  'scores':[80,70,40]})
```

```python
pd.merge(df1,df2, on='ID', how='outer')
```

```python
pd.merge(df1,df2, on='ID', how='left')
```

```python
pd.merge(df1,df2, on='ID', how='right')
```

```python
pd.merge(df1,df2, on='ID', how='inner')
```

```python
df3=pd.DataFrame({'ID_COLUMN':[1,2,3,4],
                  'score':[40,80,60,30]})
```

```python
pd.merge(df1,df3,left_on='ID',right_on='ID_COLUMN',how='outer')
```

```python
df1.join(df2,on='ID',lsuffix='left',rsuffix='right',how='inner')
```

```python
df1_index=df1.set_index('ID')
df2_index=df2.set_index('ID')
```

```python
df1_index.join(df2_index,how='inner')
```

```python
pd.concat([df1,df2],axis=0,ignore_index=True)
```

```python
pd.concat([df1,df2],axis=1,ignore_index=True)
```
