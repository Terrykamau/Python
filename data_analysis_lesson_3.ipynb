```python
import pandas as pd
import numpy as np
```

```python
data={'name': ['teresiah','Njoki','kamau','wambui','Mary'],
      'country':['kenya','uganda',1,'djibouti',np.nan],
      'age':[20,30,40,np.nan,60]}
df=pd.DataFrame(data)
print(df)
```

```python
df['age'].ffill()
```

```python
df.dtypes
```

```python
df['country'].apply(type).value_counts()
```

```python
df['Range']=df['age'].apply(lambda age:'x'if age<25 else "")
```

```python
df['Z_score']=df['age']-df['age'].mean()/df['age'].std()
```

```python
df['Z_score']=df['Z_score'].apply(lambda value:"x" if value <-2 or value > 2 else "")
```

```python
q1=df['age'].quantile(q=0.25)
q3=df['age'].quantile(q=0.75)
iqr=q3-q1
print(iqr)
```

```python
lower_bound= q1-1.5*iqr
print(lower_bound)
```

```python
upper_bound=q3+1.5*iqr
print(upper_bound)
```

```python
df['iqr']=df['age'].apply(lambda age:'x' if age<lower_bound or age>upper_bound else " ")
print(df)
```

```python
from sklearn.ensemble import IsolationForest as IS
```

```python
model=IS(contamination='auto')
```

```python
df['IsolationForest']=model.fit_predict(df[['age']])
print(df)
```
