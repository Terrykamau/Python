```python
import pandas as pd
import numpy as np
```

```python
data={'name': ['teresiah','Njoki','kamau','wambui'],
      'country':['kenya','uganda','rwanda','djibouti'],
      'age':[20,30,40,np.nan]}
df=pd.DataFrame(data)
print(df)
```

```python
df.isnull().values.any()
```

```python
df.isnull().sum().sum()
```

```python
df.isnull().sum()
```

```python
df.isnull()
```

```python
df.isnull().any(axis=0)
```

```python
df.isnull().any(axis=1)
```

```python
df.dropna(axis=1, how='all')
```

```python
df.dropna(axis=0, how='all')
```

```python
df[df['age'].notnull()]
```

```python
df.fillna(0)
```

```python
df.fillna(df['age'].mean())
```

```python
df.fillna(df['age'].count())
```
