```python
import seaborn as sns
```

```python
sns.get_dataset_names()
```

```python
flights=sns.load_dataset('flights')
print(flights)
```

```python
sns.heatmap(flights.pivot(index='year',columns='month',values='passengers'),cmap='crest')
```

```python
sns.histplot(data=flights,x='passengers',bins=30,kde=True)
```

```python
sns.histplot(data=flights,x='passengers',bins=30, hue='month',multiple='stack')
```

```python
sns.boxplot(data=flights,y='passengers')
```

```python
sns.boxplot(data=flights,y='passengers',x='month')
```

```python
sns.boxplot(data=flights,y='passengers',x='year')
```

```python
sns.violinplot(data=flights,y='passengers',x='year')
```
