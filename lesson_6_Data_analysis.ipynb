```python
import pandas as pd
import numpy as np
```

```python
df=pd.DataFrame({'A':['for','for','for','bar','bar'],
                 'B':['one','one','two','two','one'],
                 'D':[1,2,2,3,3,],
                 'E':[2,3,5,6,6,],
                 'F':[0,1,2,3,4]
                 })
```

```python
print(df)
```

```python
df.melt(
    id_vars=['A'],
    var_name='Metrics',
    value_name='values'
)
```

```python
df.melt(
    id_vars=['A','B'],
    var_name='Metrics',
    value_name='values'
)
```

```python
df.stack()
```

```python
df.stack(level=0)
```

```python
df.stack().unstack()
```

```python
df.drop_duplicates()
```

```python
df.pivot(
    index='A',
    columns='F',
    values='E'
)
```

```python
df.pivot_table(
    values='E',
    index=['A','B'],
    columns='F',
    aggfunc=['sum','mean','median']
)
```
