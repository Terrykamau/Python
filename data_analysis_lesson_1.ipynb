```python
mean=df['age'].mean()
median=df['age'].median()
variance= df['age'].var()
standard_deviation=df['age'].std()
min_max=df['age'].min(), df['age'].max()
print(mean)
print(median)
print(variance)
print(standard_deviation)
print(min_max)
```
